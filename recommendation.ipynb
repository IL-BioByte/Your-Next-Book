```python
def clean_data(row, mean_values):
    if pd.isna(row['Rating']):
        row['Rating'] = mean_values['Rating']
    if pd.isna(row['Score']):
        row['Score'] = mean_values['Score']
    if pd.isna(row['Number of Votes']):
        row['Number of Votes'] = mean_values['Number of Votes']
        
    if pd.isna(row['description']):
        row['description'] = f"No description available for {row['Book Name']}"
    if pd.isna(row['categories']):
        row['categories'] = f"Unknown categories for {row['Book Name']}"
    if pd.isna(row['published_year']):
        row['published_year'] = f"Unknown published year for {row['Book Name']}"
        
    return row


```

```python
mean_values = {
    'Rating': df['Rating'].mean(),
    'Score': df['Score'].mean(),
    'Number of Votes': df['Number of Votes'].mean(),
}
```

```python
df = df.apply(lambda row: clean_data(row, mean_values), axis=1)
df.head()
```
